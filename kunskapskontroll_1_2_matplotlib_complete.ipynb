### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure - A figure, a window, a screen. A place where your get all your data drawn onto, like a canvas. All inputted data will be visualized on an area.

Axes - Represent an individual plot or graph where you visualize your data, containing elements like line, points, text. Each axes has its own set or coordinates.

Axis - responsible for setting the scale of the dimension of the plot. refers to either X-axis or Y-axis. also generates ticks and ticklabels.

Artist - Artist is everything you can see and read on a plot. for example, titles, legends, text, markers, lines, figure, axes, axis etc.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#Object oriented style
x = np.linspace(-4, 4, 100)

fig, ax= plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x**2, label='y = x^2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title("Object oriented style")
ax.legend()
```

```python
#pyplot-style
x = np.linspace(-4, 4, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x**2, label='y = x^2')  # etc.
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("pyplot-style")
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
def scatter_generator(ax, data1, data2, line_name=None):
    out = ax.scatter(data1, data2, s=3, label=line_name)
    return out

#TESTING
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
scatter_generator(ax1, random_data_x, random_data_y)
scatter_generator(ax2, x, y)
```

```python
#TESTING
fig, ax3 = plt.subplots(figsize=(5, 2.7), layout='constrained')

ax3.bar(names, values)
```

```python
#ANSWER

fig, (ax4, ax5) = plt.subplots(1, 2, figsize=(15, 4))
fig.suptitle("Differents plots", size = 30)

scatter_generator(ax4, random_data_x, random_data_y, "random_data")
scatter_generator(ax4, x, y, "y = x^2")
ax4.set_xlabel('X-axis')
ax4.set_ylabel('Y-axis')
ax4.set_title("Scatter plot")
ax4.legend()


ax5.bar(names, values)
ax5.set_xlabel('Fruits')
ax5.set_ylabel('Value')
ax5.set_title("Bar plot")
```
